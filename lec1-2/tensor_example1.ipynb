```python
import numpy as np
import matplotlib.pyplot as plt

plate = np.array([
    [[255,   0,   0], [  0, 255,   0], [  0,   0, 255]],
    [[255, 255,   0], [  0, 255, 255], [255,   0, 255]],
], dtype=np.uint8)

R_only = plate * np.array([1,0,0], dtype=np.uint8)
G_only = plate * np.array([0,1,0], dtype=np.uint8)
B_only = plate * np.array([0,0,1], dtype=np.uint8)

plt.figure(figsize=(10,3))
for i,(img,title) in enumerate([(plate,"RGB"), (R_only,"R only"), (G_only,"G only"), (B_only,"B only")], 1):
    plt.subplot(1,4,i); plt.title(title); plt.imshow(img, interpolation="nearest"); plt.axis("off")
plt.tight_layout(); plt.show()
```
