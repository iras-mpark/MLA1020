```python
import numpy as np
import matplotlib.pyplot as plt

plate = np.array([
    [[255,   0,   0], [  0, 255,   0], [  0,   0, 255]],
    [[255, 255,   0], [  0, 255, 255], [255,   0, 255]],
], dtype=np.uint8)

R_only = plate * np.array([1,0,0], dtype=np.uint8)
G_only = plate * np.array([0,1,0], dtype=np.uint8)
B_only = plate * np.array([0,0,1], dtype=np.uint8)

plt.figure(figsize=(10,3))
for i,(img,title) in enumerate([(plate,"RGB"), (R_only,"R only"), (G_only,"G only"), (B_only,"B only")], 1):
    plt.subplot(1,4,i); plt.title(title); plt.imshow(img, interpolation="nearest"); plt.axis("off")
plt.tight_layout(); plt.show()
```

# General Tensor form in AI

```python
# A D-dimensional data point
D = 2  # Number of dimensions
x = np.ones(D)
x
```

```python
N = 3  # Number of examples
# A dataset of N examples, each D-dimensional point
x = np.ones((N, D))
x
```

```python
L = 4  # Length of sequence
# A dataset of N examples, each length L, each position is D-dimensional
x = np.ones((N, L, D))
x
```

```python
H = 2  # A height
W = 2  # A width
C = 3  # A number of channels (red, green, blue)
x = np.ones((N, H, W, C))
x
```

```python
Din = 3  # ...a Din-dimensional input
Dout = 2  # ...to a Dout-dimensional output
w = np.ones((Din, Dout))
w
```

# Tensor Slicing, Transpose

```python
x = np.random.randn(2, 3)
x
```

```python
# Get row 0
y = x[0]
y
```

```python
# Get column 1
y = x[:, 1]
y
```

```python
# Transpose the matrix
y = x.transpose(1, 0)
y
```

```python
# Transpose는 이미 생성된 차원 내에서만 활용 가능
z = np.array([1,2,3,4,5])
z
```

```python
# (5,)를 (1,5)로 변경 원함
z.transpose(0,1) # Error
```
