# EinOps methods (p. 27)

### einsum

```python
import torch
import einops

x = torch.ones(3, 4)  # seq1 hidden
y = torch.ones(4, 3)  # hidden seq2
# Old way
z = x @ y   # seq1 seq2
z
```

```python
# New (einops) way
z = einops.einsum(x, y, "seq1 hidden, hidden seq2 -> seq1 seq2")
z
```

```python
x = torch.ones(2, 3, 4)  # batch seq1 hidden
y = torch.ones(2, 3, 4)  # batch seq2 hidden
# Old way
z = x @ y.transpose(-2, -1)  # batch seq1 seq2
z
```

```python
z = einops.einsum(x, y, "batch seq1 hidden, batch seq2 hidden -> batch seq1 seq2")
z
```

```python
# 계산에 영향 적은 부분 생략 가능
z = einops.einsum(x, y, "... seq1 hidden, ... seq2 hidden -> ... seq1 seq2")
z
```

### reduce

```python
x = torch.ones(2, 3, 4)  # batch seq hidden
# Old way
y = x.sum(dim=-1)
y
```

```python
# New (einops) way
y = einops.reduce(x, "... hidden -> ...", "sum")
y
```

### rearrange

```python
x = torch.ones(3, 8)  # seq total_hidden
x
```

```python
# We want to total_hidden -> heads * hidden = (2) * (4)
x = einops.rearrange(x, "... (heads hidden1) -> ... heads hidden1", heads=2)
x
```

```python
# 재배열된 x를 가중치와 행렬곱 (사람이 읽기 좋은 형태)
w = torch.ones(4, 4)  # hidden1 hidden2
x = einops.einsum(x, w, "... hidden1, hidden1 hidden2 -> ... hidden2")
x
```

```python
# 다시 (3,8) 형태로 변경도 쉬움
x = einops.rearrange(x, "... heads hidden2 -> ... (heads hidden2)")
x
```
